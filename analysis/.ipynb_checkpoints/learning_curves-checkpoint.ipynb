```python
import os
import json
import numpy as np
import matplotlib.pyplot as plt
```

```python
RES = os.path.join('..', 'res')
```

```python
FEVAL_SARNN = os.path.join(RES, 'feval-overall-sarnn-1540522028.json')
FEVAL_LSTM = os.path.join(RES, 'feval-overall-lstm-1540388590.json')
```

```python
lines = []
with open(FEVAL_SARNN, 'r') as f:
    for line in f:
        line = line.strip()
        line = eval(line)
        if 'Metrics' not in line:
            continue
        lines.append(line)
        
acc = []
loss = []
grad = []
for i, line in enumerate(lines):
    a, l, g = line['Metrics']
    acc.append(a)
    loss.append(l)
    grad.append(g)
loss = np.array(loss) / max(loss)
grad = np.array(grad) / max(grad)
```

```python
len(lines)
```

```python
nepochs = len(lines)/4
epoch = np.arange(0, nepochs, 0.25)
```

```python
plt.plot(epoch, acc,
        label='acc')
plt.plot(epoch, loss,
        label='loss')
plt.plot(epoch, grad,
        label='grad')
plt.legend()
plt.xlabel('#epoch')
plt.title('learning curves of sarnn on feval')
# plt.xticks(np.arange(0, seq_len, 1), list(seq))
# plt.yticks(np.arange(0, 1.1, 0.1))
plt.show()
```
