```python
def mem_stack():
    kept = 9
    remain = 3
    seq = '8 + 6 * 3 / 2 - 4 * 7 * 4 / 5 - 1 / 5 / 4 * 9 / 9 + 1 + 8 * 4 - 7 * 9 * 9 + 3 / 4 + 7 / 9 - 6 / 6 * 4 + 3 / 3'.split(' ')
    seq[:9] = [r'$\langle \mathrm {N_0} \rangle$',
               r'$\langle \mathrm {L_0} \rangle$',
               r'$\langle \mathrm {N_1} \rangle$',
               r'$\langle \mathrm {H_0} \rangle$',
               r'$\langle \mathrm {N_2} \rangle$',
               r'$\langle \mathrm {H_1} \rangle$',
               r'$\langle \mathrm {N_3} \rangle$',
               r'$\langle \mathrm {L_1} \rangle$',
               r'$\langle \mathrm {N_4} \rangle$']
    seq_len = len(seq)
    ds = '1 5 1 2 3 4 6 7 1 2 3 4 5 6 8 9 1 2 3 4 5 6 7 8 10 11 12 13 1 2 14 15 1 2 3 4 16 17 1 2 18 19 1 2 20 21 1 2 3 4 22 23 1 2'.split(' ')
    syn_dis = [int(d) for d in ds]
    
    prob_action = {i:[] for i in range(seq_len)}
    mem = {i:[] for i in range(seq_len)}
    i = 0
    nc = 0
    nt = 0
    with open('feval-analy_d23_ne10-sarnn.txt', 'r') as f:
        for line in f:
            line_json = json.loads(line)
            if line_json['type'] != 'actions':
                assert line_json['type'] == 'input'
                nc += line_json['is_correct']
                nt += 1
                continue 

            prob_action[i].append(line_json['all'])
            mem[i].append(line_json['mem'])        
            i += 1
            i %= seq_len
    print('correct prediction%:', nc/nt)

    step = []
    for i in range(seq_len):
        step.append(np.array(mem[i]).mean(axis=0)[:, :remain])
    mean_mem_dist = []
    for i in range(kept):
        mean_mem_dist.append(np.hstack([step[i], np.zeros((5, 1))-0.2]))
    mem_whole = np.hstack(mean_mem_dist)[:, :-1]
    mem_whole.shape
    mem_whole = mem_whole
    ncols = mem_whole.shape[1]
    fig, ax = plt.subplots()
    fig.dpi = 200
    im = ax.imshow(mem_whole, cmap=plt.cm.Blues)
#     ax.figure.colorbar(im, ax=ax, shrink=0.3)
    plt.xticks(np.arange(1, ncols, 4), seq[:kept], fontsize=10, family='TImes New Roman')
    plt.yticks(np.arange(5), fontsize=10, family='Times New Roman')
    plt.xlabel('input symbols along time', family='TImes New Roman', fontsize=10)
    plt.ylabel('address', family='Times New Roman', fontsize=10)
    plt.show()
```

```python
def sarnn_policy():
    seq = '8 + 6 * 3 / 2 - 4 * 7 * 4 / 5 - 1 / 5 / 4 * 9 / 9 + 1 + 8 * 4 - 7 * 9 * 9 + 3 / 4 + 7 / 9 - 6 / 6 * 4 + 3 / 3'.split(' ')
    seq[:9] = [r'$\langle \mathrm {N_0} \rangle$',
               r'$\langle \mathrm {L_0} \rangle$',
               r'$\langle \mathrm {N_1} \rangle$',
               r'$\langle \mathrm {H_0} \rangle$',
               r'$\langle \mathrm {N_2} \rangle$',
               r'$\langle \mathrm {H_1} \rangle$',
               r'$\langle \mathrm {N_3} \rangle$',
               r'$\langle \mathrm {L_1} \rangle$',
               r'$\langle \mathrm {N_4} \rangle$']
    seq_len = len(seq)
    ds = '1 5 1 2 3 4 6 7 1 2 3 4 5 6 8 9 1 2 3 4 5 6 7 8 10 11 12 13 1 2 14 15 1 2 3 4 16 17 1 2 18 19 1 2 20 21 1 2 3 4 22 23 1 2'.split(' ')
    syn_dis = [int(d) for d in ds]
    kept = 9
    prob_action = {i:[] for i in range(seq_len)}
    i = 0
    nc = 0
    nt = 0
    with open('feval-analy_d23_ne10-sarnn.txt', 'r') as f:
        for line in f:
            line_json = json.loads(line)
            if line_json['type'] != 'actions':
                assert line_json['type'] == 'input'
                nc += line_json['is_correct']
                nt += 1
                continue 

            prob_action[i].append(line_json['all'])
            i += 1
            i %= seq_len
    step = []
    for i in range(seq_len):
        step.append(np.array(prob_action[i]))
    mean_dist = []
    for i in range(seq_len):
        mean_dist.append(step[i].sum(0)/step[i].shape[0])
    N = 5 
    push = [mean_dist[t][:N+1].sum() for t in range(seq_len)]
    npop = []
    for t in range(seq_len):
        pop = (mean_dist[t][:N+1] * np.arange(N+1)).sum() +\
            (mean_dist[t][N+1:] * np.arange(N+1)).sum()
        npop.append(pop)    
    pop_normalized = np.array(npop) / (N+1)
    kept = 9
    fig, ax = plt.subplots()
    fig.dpi = 100
    plt.grid(axis='both', linestyle='dashed', linewidth=0.5)
#     lns0 = ax.bar(np.arange(0.5, seq_len-0.5)[:kept-1], (np.array(syn_dis)/max(syn_dis))[:kept-1],
#              width=0.5, label='syn-distance', color='gray', edgecolor='black')
    lns1 = ax.plot(np.arange(seq_len)[:kept], 
             push[:kept],
            label='push', marker='s',color='blue', alpha=0.5, fillstyle='none')
    ax.set_ylabel('Push probability', fontsize=15, family='TImes New Roman')
    ax.set_yticks(np.arange(0, 1+0.1, 0.1))
    ax.set_yticklabels([round(x, 1) for x in np.arange(0, 1+0.1, 0.1)], fontdict={'family': 'TImes New Roman', 'size': 15})
    ax.set_xlabel('input symbols along time', fontsize=15, family='TImes New Roman')
    
    ax_r = ax.twinx()
    lns2 = ax_r.plot(np.arange(seq_len)[:kept], 
         npop[:kept],
        label='pop', marker='o', color='red', alpha=0.5, fillstyle='none')
    ax_r.set_ylabel('Expectation of #pop', fontsize=15, family='TImes New Roman')
    ax_r.set_yticks(np.arange(0, 5.5, 0.5))
    ax_r.set_yticklabels(np.arange(0, 5.5, 0.5), fontdict={'family': 'TImes New Roman', 'size': 15})

#     lns = [lns0] + lns1 + lns2
    lns = lns1 + lns2
    labs = [l.get_label() for l in lns]
    plt.legend(lns, labs, loc=0, framealpha=0, prop={'family': 'TImes New Roman', 'size': 15})
    plt.xticks(np.arange(0, seq_len, 1)[:kept])
    ax.set_xticklabels(list(seq)[:kept], fontdict={'family': 'TImes New Roman', 'size': 15})
    plt.show()
```

```python
def mem_ntm(idx, kept=9):
    seq = '8 + 6 * 3 / 2 - 4 * 7 * 4 / 5 - 1 / 5 / 4 * 9 / 9 + 1 + 8 * 4 - 7 * 9 * 9 + 3 / 4 + 7 / 9 - 6 / 6 * 4 + 3 / 3'.split(' ')
    seq_len = len(seq)
    ds = '1 5 1 2 3 4 6 7 1 2 3 4 5 6 8 9 1 2 3 4 5 6 7 8 10 11 12 13 1 2 14 15 1 2 3 4 16 17 1 2 18 19 1 2 20 21 1 2 3 4 22 23 1 2'.split(' ')
    syn_dis = [int(d) for d in ds]
    wr = {i:[] for i in range(seq_len)}
    ww = {i:[] for i in range(seq_len)}
    es = []

    i = 0
    nc = 0
    nt = 0
    with open('feval-analy_d23_ne10-ntm.txt', 'r') as f:
    # with open('feval-change_numerals-ntm-1544630154.txt', 'r') as f:
        for line in f:
            line_json = json.loads(line)
            if 'w' not in line_json:
                nc += line_json['is_correct']
                nt += 1
                e = line_json['expr']
                es.append(e)
                continue

            if line_json['type'] == 'read':
                wr[i].append(line_json['w'])
            if line_json['type'] == 'write':
                ww[i].append(line_json['w'])
            i += 1
            i %= seq_len
    
    wr_temp = {i:[] for i in range(seq_len)}
    ww_temp = {i:[] for i in range(seq_len)}

    for i in range(seq_len):
        wr_temp[i] = np.array(wr[i][idx:idx+1]).sum(0)/len(wr[i][idx:idx+1])
        ww_temp[i] = np.array(ww[i][idx:idx+1]).sum(0)/len(ww[i][idx:idx+1])
    wrs = np.array([wr_temp[i] for i in range(seq_len)])
    wws = np.array([ww_temp[i] for i in range(seq_len)])
    ws = np.hstack([wrs, np.zeros((seq_len, 1)), wws])
    fig, ax = plt.subplots()
    fig.dpi = 100
#     plt.title('read(top) and write(bottom)', fontsize=10)
    im = ax.imshow(ws.transpose()[:, :kept], cmap=plt.cm.Blues)
#     plt.grid(linestyle='dashed', linewidth=0.1)
    ax.figure.colorbar(im, ax=ax)
    plt.yticks(np.arange(0, 11), [str(i) for i in range(5)] + [''] + [str(i) for i in range(5)], fontsize=15)
    plt.xticks(np.arange(0, seq_len)[:kept], list(es[idx])[:kept], fontsize=15)
    plt.ylabel('memory cell position', fontsize=15)
    plt.xlabel('input symbols along time', fontsize=10)
#     plt.title('top read, bottom write')
```

```python
def mem_ntm_sep(idx, kept=9):
    seq = '8 + 6 * 3 / 2 - 4 * 7 * 4 / 5 - 1 / 5 / 4 * 9 / 9 + 1 + 8 * 4 - 7 * 9 * 9 + 3 / 4 + 7 / 9 - 6 / 6 * 4 + 3 / 3'.split(' ')
    seq_len = len(seq)
    ds = '1 5 1 2 3 4 6 7 1 2 3 4 5 6 8 9 1 2 3 4 5 6 7 8 10 11 12 13 1 2 14 15 1 2 3 4 16 17 1 2 18 19 1 2 20 21 1 2 3 4 22 23 1 2'.split(' ')
    syn_dis = [int(d) for d in ds]
    wr = {i:[] for i in range(seq_len)}
    ww = {i:[] for i in range(seq_len)}
    es = []

    i = 0
    nc = 0
    nt = 0
    with open('feval-analy_d23_ne10-ntm.txt', 'r') as f:
    # with open('feval-change_numerals-ntm-1544630154.txt', 'r') as f:
        for line in f:
            line_json = json.loads(line)
            if 'w' not in line_json:
                nc += line_json['is_correct']
                nt += 1
                e = line_json['expr']
                es.append(e)
                continue

            if line_json['type'] == 'read':
                wr[i].append(line_json['w'])
            if line_json['type'] == 'write':
                ww[i].append(line_json['w'])
            i += 1
            i %= seq_len
    
    wr_temp = {i:[] for i in range(seq_len)}
    ww_temp = {i:[] for i in range(seq_len)}

    for i in range(seq_len):
        wr_temp[i] = np.array(wr[i][idx:idx+1]).sum(0)/len(wr[i][idx:idx+1])
        ww_temp[i] = np.array(ww[i][idx:idx+1]).sum(0)/len(ww[i][idx:idx+1])
    wrs = np.array([wr_temp[i] for i in range(seq_len)])
    wws = np.array([ww_temp[i] for i in range(seq_len)])
    
    
#     ws = np.hstack([wrs, np.zeros((seq_len, 1)), wws])
    fig, ax = plt.subplots()
    fig.dpi = 100
#     plt.title('read(top) and write(bottom)', fontsize=10)
#     im = ax.imshow(ws.transpose()[:, :kept], cmap=plt.cm.Blues)
    im = ax.imshow(wrs.transpose()[:, :kept], cmap=plt.cm.Blues)
#     plt.grid(linestyle='dashed', linewidth=0.1)
    cbar = ax.figure.colorbar(im, ax=ax, shrink=0.7)
    cbar.ax.tick_params(labelsize=15)
    plt.yticks(np.arange(0, 5), [str(i) for i in range(5)], fontsize=15,family='Times New Roman')
    plt.xticks(np.arange(0, seq_len)[:kept], list(es[idx])[:kept], fontsize=15,family='Times New Roman')
    plt.ylabel('address', fontsize=15, family='Times New Roman')
    plt.xlabel('input symbols along time', fontsize=15,family='Times New Roman')
#     plt.title('read policy', fontsize=15,family='Times New Roman')
    plt.show()
    
    fig, ax = plt.subplots()
    fig.dpi = 100
    im = ax.imshow(wws.transpose()[:, :kept], cmap=plt.cm.Blues)
#     plt.grid(linestyle='dashed', linewidth=0.1)
    cbar = ax.figure.colorbar(im, ax=ax, shrink=0.7)
    cbar.set_ticks(np.arange(0, 1.2, 0.2))
    plt.yticks(np.arange(0, 5), [str(i) for i in range(5)], fontsize=15,family='Times New Roman')
    plt.xticks(np.arange(0, seq_len)[:kept], list(es[idx])[:kept], fontsize=15,family='Times New Roman')
    plt.ylabel('address', fontsize=15,family='Times New Roman')
    plt.xlabel('input symbols along time', fontsize=15,family='Times New Roman')
#     plt.title('write policy', fontsize=15,family='Times New Roman')
    plt.show()
```

```python
def sarnn_policy_pa1mmc3():
    seq = ['8', '+', '6', '*', '3', '/', '(', '2', '-', '4', ')']
    syn_dis = [1, 6, 1, 2, 3, 5, 3, 1, 2, 4]
    fpath = 'feval-analysis-sarnn-1545981241.txt'
    prob_action = {i:[] for i in range(seq_len)}
    
    kept=9
    i = 0
    nc = 0
    nt = 0
    with open(fpath, 'r') as f:
        for line in f:
            line_json = json.loads(line)
            if line_json['type'] != 'actions':
                assert line_json['type'] == 'input'
                nc += line_json['is_correct']
                nt += 1
                continue 

            prob_action[i].append(line_json['all'])
            i += 1
            i %= seq_len   
        
    step = []
    for i in range(seq_len):
        step.append(np.array(prob_action[i]))
    mean_dist = []
    for i in range(seq_len):
        mean_dist.append(step[i].sum(0)/step[i].shape[0])
    push = [mean_dist[t][:11].sum() for t in range(seq_len)]
    npop = []
    for t in range(seq_len):
        pop = (mean_dist[t][:11] * np.arange(11)).sum() + \
            (mean_dist[t][11:] * np.arange(11)).sum()
        npop.append(pop)    
    npop = []
    for t in range(seq_len):
        pop = (mean_dist[t][:11] * np.arange(11)).sum() + \
            (mean_dist[t][11:] * np.arange(11)).sum()
        npop.append(pop)    
        
    fig, ax = plt.subplots()
    fig.dpi = 200
    plt.grid(axis='both', linestyle='dashed', linewidth=0.5)
    lns0 = ax.bar(np.arange(0.5, seq_len-0.5)[:kept-1], (np.array(syn_dis)/max(syn_dis))[:kept-1]*0.3,
            width=0.5, label='syn-distance', color='gray', edgecolor='black')

    lns1 = ax.plot(np.arange(seq_len)[:kept], 
             push[:kept],
            label='push', marker='s',color='blue', alpha=0.5, fillstyle='none')
    ax.set_ylabel('Push probabiliy', fontsize=10)
    ax.set_yticks(np.arange(0, 0.5+0.1, 0.1))
    ax.set_xlabel('input symbols along time', fontsize=10)
    
    ax_r = ax.twinx()
    lns2 = ax_r.plot(np.arange(seq_len)[:kept], 
             npop[:kept],
            label='pop', marker='o', color='red', alpha=0.5, fillstyle='none')
    ax_r.set_ylabel('Expectation of #pop', fontsize=10)
    ax_r.set_yticks(np.arange(0, 10+1, 1))

    lns = [lns0] + lns1 + lns2
    labs = [l.get_label() for l in lns]
    plt.legend(lns, labs, loc=0, fontsize=10)

#     plt.xlabel('input symbol')

    #     plt.legend()
    plt.xticks(np.arange(0, seq_len, 1)[:kept], list(seq)[:kept], fontsize=15)
    # plt.yticks(np.arange(0, 1.1, 0.1), fontsize=10)
    plt.show()
```
