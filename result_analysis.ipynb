```python
def draw_chart(res_dict):
    fig, ax = plt.subplots()
    fig.dpi = 200
    markers = ['o', 'v', 's', 'p', '*']
    for i, (mname, res) in enumerate(res_dict.items()):
        x, y = [], []
        for h, acc in res:
            x.append(h)
            y.append(acc)
        plt.plot(x, y, label=mname, marker=markers[i])
    plt.legend()
    plt.grid()
    plt.grid(color='b' , linewidth=0.5 ,linestyle='--')
    plt.yticks(np.arange(0,1.1,0.1))
    plt.ylabel('accuracy')
    plt.xlabel('depth')
    plt.show()
    
```
